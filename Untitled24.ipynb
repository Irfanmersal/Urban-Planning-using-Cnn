```python
import matplotlib.pyplot as plt
import cv2
def cv2_imshow(a, **kwargs):
    a = a.clip(0, 255).astype('uint8')
    # cv2 stores colors as BGR; convert to RGB
    if a.ndim == 3:
        if a.shape[2] == 4:
            a = cv2.cvtColor(a, cv2.COLOR_BGRA2RGBA)
        else:
            a = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)

    return plt.imshow(a, **kwargs)
```

```python
import cv2
from google.colab import files
import matplotlib.pyplot as plt

# Upload an image from your local machine
uploaded = files.upload()

# Load the uploaded image
image = cv2.imread(list(uploaded.keys())[0])

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection
edges = cv2.Canny(gray_image, threshold1=30, threshold2=100)  # You can adjust the thresholds as needed

# Display the original, grayscale, and edge-detected images
plt.figure(figsize=(15, 5))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Grayscale Image
plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Edge-Detected Image
plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Edge-Detected Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the satellite image (you need to upload an image)
from google.colab import files
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
satellite_image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(satellite_image, cv2.COLOR_BGR2GRAY)

# Threshold the grayscale image to separate urban areas from non-urban areas
# Adjust the threshold value based on your image
_, thresholded_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)

# Perform morphological operations (e.g., dilation) to enhance urban areas
kernel = np.ones((5, 5), np.uint8)
enhanced_image = cv2.dilate(thresholded_image, kernel, iterations=2)

# Find contours of urban areas
contours, _ = cv2.findContours(enhanced_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Create a mask for urban areas
urban_mask = np.zeros_like(gray_image)
cv2.drawContours(urban_mask, contours, -1, (255), thickness=cv2.FILLED)

# Calculate the area of urban sprawl (in pixels) and total area (in pixels)
urban_area = np.sum(urban_mask == 255)
total_area = urban_mask.size

# Calculate the percentage of urban sprawl
urban_sprawl_percentage = (urban_area / total_area) * 100

# Display the urban sprawl mask and statistics
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(urban_mask, cmap='gray')
plt.title('Urban Sprawl Mask')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.text(0.1, 0.5, f'Urban Sprawl Percentage: {urban_sprawl_percentage:.2f}%', fontsize=12, transform=plt.gca().transAxes)
plt.axis('off')

plt.show()
```
